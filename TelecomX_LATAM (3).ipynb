#ðŸ”§ TransformaciÃ³n

```python
df_customer = datos['customer'].apply(pd.Series)
df_customer.head()
```

```python
# Expandir cada columna JSON en columnas separadas
df_customer = datos['customer'].apply(pd.Series)
df_phone = datos['phone'].apply(pd.Series)
df_internet = datos['internet'].apply(pd.Series)
df_account = datos['account'].apply(pd.Series)

# Unir todo en un solo DataFrame
df_final = pd.concat([datos.drop(['customer', 'phone', 'internet', 'account'], axis=1),
                      df_customer, df_phone, df_internet, df_account], axis=1)

print(df_final)
```

```python
df_final.head()
```

```python
df_charges = df_final['Charges'].apply(pd.Series)
```

```python
df_final2 = pd.concat([df_final.drop(['Charges'], axis=1),
                      df_charges], axis=1)
df_final2.head()
```

```python
df_final2['Partner'] = df_final2['Partner'].map({'Yes': 1, 'No': 0})
df_final2['Dependents'] = df_final2['Dependents'].map({'Yes': 1, 'No': 0})
df_final2['PhoneService'] = df_final2['PhoneService'].map({'Yes': 1, 'No': 0})
df_final2['PaperlessBilling'] = df_final2['PaperlessBilling'].map({'Yes': 1, 'No': 0})
df_final2['Churn'] = df_final2['Churn'].map({'Yes': 1, 'No': 0})
df_final2['MultipleLines'] = df_final2['MultipleLines'].map({'Yes': 1, 'No': 0})
df_final2['OnlineSecurity'] = df_final2['OnlineSecurity'].map({'Yes': 1, 'No': 0})
df_final2['OnlineBackup'] = df_final2['OnlineBackup'].map({'Yes': 1, 'No': 0})
df_final2['DeviceProtection'] = df_final2['DeviceProtection'].map({'Yes': 1, 'No': 0})
df_final2['TechSupport'] = df_final2['TechSupport'].map({'Yes': 1, 'No': 0})
df_final2['StreamingTV'] = df_final2['StreamingTV'].map({'Yes': 1, 'No': 0})
df_final2['StreamingMovies'] = df_final2['StreamingMovies'].map({'Yes': 1, 'No': 0})

```
